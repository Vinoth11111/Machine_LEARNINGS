```python
# l2 ridge - lamda * summation of w^2
# we are adding 1/2 for scaling and 2 in mse gradient is cancelled out
# square from l2 came before and canceled with 2 finally we got gradient. = 1/m *(x*(error)) + lamb
```

```python
x = (4+np.random.randint(1,100,100))-2
noice = np.random.normal(1,15,100)
y = 4*x +2+noice
```

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def Regularized_Regression(x,y,epoch,lr,lamb):
  m = 0
  c = 0
  for _ in range(epoch):
    y_pred = m*x + c
    error = y-y_pred
    reg_grad_m = ((1/len(x))*x.dot(error)) + ((lamb/len(x))*m)
    reg_grad_c = (1/len(x)*np.sum(error)) + ((lamb/len(x))*c)
    up_m = lr*reg_grad_m
    up_c = lr*reg_grad_c
    if np.any(np.isnan(up_c)) or np.any(np.isnan(up_m)) or np.any(np.isinf(up_c)) or np.any(np.isinf(up_m)):
      break
    m += up_m
    c += up_c
  return m,c,y_pred
```

```python
def scale(x):
  mean = np.mean(x)
  std = np.std(x)
  scale = (x-mean)/std
  return scale
```

```python
m,c,y_pred = Regularized_Regression(scale(x),y,10000,0.001,0.1)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color='r',label = 'Predicted Line')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
m,c,y_pred = Regularized_Regression(scale(x),y,10000,0.001,1)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color='r',label = 'Predicted Line')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
