```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic_regression(x,y,lr,epochs):
  def sigmoid(z):
    return 1/(1+ np.exp(-z))

  w = np.zeros(x.shape[1]) # ceating co-efficient for weight
  b = 0 # initialling bias

  for i in range(epochs):
    z = np.dot(x,w) + b # here we are adding bias to the weight directly without creating new columns in the feature matrix.
    y_pred = sigmoid(z)
    if i % 5 == 0: # printing loss at every 5th steps.
      loss = (-1/len(x))*np.sum((y*np.log(y_pred) + (1-y)*np.log(1-y_pred)))
      print(f'loss at epoch {i} is {loss}')
    error = y-y_pred #cal error for gradient
    gw = -(1/len(x))*np.dot(error,x) #finding gradient for weight
    gb = -(1/len(x))*np.sum(error) #gradient for bias
    if np.any(np.isnan(gw)) or np.any(np.isnan(gb)) or np.any(np.isinf(gw)) or np.any(np.isinf(gb)): #for handling nan and infinte error
      break
    w += lr*gw #updating weight
    b += lr*gb #updating bias
  return w,b #returning weight and bias
```

```python
X = np.array([[0.1, 0.2], [0.4, 0.5], [0.7, 0.8]])  # Feature matrix (3 samples, 2 features)
y = np.array([0, 1, 1])
```

```python
w,b = logistic_regression(X,y,0.01,100)
```

Since the input is very small, 3,2 we have high loss, we can reduce the epoches or incrase the learning rate.

To see the boundary of the logistic function

```python
# Compute z values
z = np.dot(X, w) + b
print("z shape:", z.shape)   # (3,)

# Plot z
plt.figure(figsize=(6,4))
plt.plot(range(len(z)), z, 'o-', label='z values')
plt.title('Linear combination (z = Xw + b)')
plt.xlabel('Sample index')
plt.ylabel('z value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def predict(x,w,b):
  z = np.dot(x,w) +b
  y_pred = 1/(1+ np.exp(-z))
  if y_pred >= 0.5:
    return 1
  else:
    return 0
```

```python
predicted = predict([[0.3,0.7]],w,b)
print(predicted)
```
