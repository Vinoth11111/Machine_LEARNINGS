```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = 4+(4+np.random.randint(1,100,100))-2
noice = np.random.normal(0,10,100)
y = 5+3*(x)+noice
```

```python
plt.scatter(x,y)
plt.title('X and Y Scatter Plot with Noice')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def LinearRegression(x,y,epochs,lr):
  m = 0 #weight
  c = 0 #intercept

  for _ in range(epochs):
    y_pred = x*m + c
    error = y-y_pred
    w_gradient =(-2/len(x))*(error.dot(x))
    c_gradient = (-2/len(x))*np.sum(error)
    w_upd = lr * w_gradient
    c_upd = lr * c_gradient
    if (np.any(np.isnan(c_upd))) or (np.any(np.isinf(c_upd))) or (np.any(np.isinf(w_upd))) or (np.any(np.isnan(w_upd))):
      break
    c -= c_upd
    m -= w_upd
  return m,c,y_pred
```

```python
m,c,y_pred = LinearRegression(x,y,500,0.005)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color = 'r',label = 'Predicted Line')
plt.legend()
plt.title('X and Y Scatter Plot with Noice')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# standadization
mean = np.mean(x)
std = np.std(x)
x_stand = (x-mean)/std
```

```python
m,c,y_pred = LinearRegression(x_stand,y,500,0.005)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color = 'r',label = 'Predicted Line')
plt.legend()
plt.title('Standardized X and Y Scatter Plot with Noice')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
