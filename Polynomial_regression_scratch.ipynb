we have to create input according to our notion(polynomial behaviour) here x in the range of -2 to 2
y is created for quatratic equation, 4+2x+5x^2+noice to get the polynomial behaviour

For plotting the prdicted line in the graph we have plot x vs predicted,here we generated value using random so we have to order the x value than plot it to see the polynomial curve

```python
def poly_features(x,degree):
  poly_x = np.ones((x.shape[0],degree+1))
  for i in range(1,degree+1):
    poly_x[:,i] = x**i
  return poly_x

def polynomial(x,y,degree,lr,epoch):
  poly_x = poly_features(x,degree)
  b = np.zeros(poly_x.shape[1])

  for _ in range(epoch):
    y_pred = poly_x.dot(b)
    error = y-y_pred
    gradient = (2/len(x))*(poly_x.T.dot(error))
    up_b = lr*gradient #actual formual for grdient is b = b- lr* grdient of the loss,
    # and in the finding the gradient grd = -2/m*(error) so both minus is canceled out i used + sign in parameter update
    # orginal formula implementation is done in the LinearRegression that there i used - sign in both gradient and param update.
    if (np.any(np.isnan(up_b)) or np.any(np.isinf(up_b))):
      break
    b += up_b
  return b, y_pred
```

```python
x = 4*np.random.rand(100,1)-2
y = 4+2*x+5*x**2+np.random.rand(100,1)
b,y_pred = polynomial(x.flatten(),y.flatten(),2,0.01,500)
```

```python
plt.scatter(x,y)
plt.title('polynomial curve')
plt.xlabel('X value(range -2 to 2)')
plt.ylabel('Y Value(5x^2 + 2x + 4)')
plt.show()
```

```python
plt.scatter(x,y)
plt.plot(x.flatten()[np.argsort(x.flatten())],y_pred[np.argsort(x.flatten())],color = 'r',label = 'predicted')
plt.legend()
plt.title('polynomial curve')
plt.xlabel('X value(range -2 to 2)')
plt.ylabel('Y Value(5x^2 + 2x + 4)')
plt.show()
```
